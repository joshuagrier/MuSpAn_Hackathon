```python
import muspan as ms
import numpy as np  
import matplotlib.pyplot as plt
## Need to install pathlib
import pathlib
```

```python
## Change to location of AKPT_Liver_Mets folder
directory = pathlib.Path("../data/everything/local_mgn_hack_data/Muspan_data/AKPT_Liver_Mets")
## Set the output directory to save the domains with the detailed met annotations
output_path = "../domains_with_mets"
```

```python
i=0
list_of_files = []
for item in directory.rglob("*"):
    if item.is_file():
            if str(item) != str(directory)+'/.DS_Store':
                list_of_files.append(str(item))
            i+=1
```

```python
names = [x.replace(str(directory)+'/', '').replace('4d/', '').replace('28d/', '').replace('.muspan', '') for x in list_of_files]
```

```python
list_of_domains = []
for i in range(len(list_of_files)):
    list_of_domains.append(ms.io.load_domain(list_of_files[i]))
```

```python
## this will take a while to run
for i in range(len(list_of_domains)):
    query_tumour = ms.query.query(list_of_domains[i], ('label', 'Broad celltype (liver)'), 'is', 'Tumour Epithelial')
    list_of_domains[i].convert_objects(population=query_tumour, 
                                collection_name = 'Tumour_Met_New',
                                object_type = 'shape',
                                conversion_method = 'alpha shape',
                                conversion_method_kwargs = dict(alpha = 30))
    ## if you want to visualise the shapes uncomment the following, but this will significantly increase the runtime
    # fig,axes=plt.subplots(figsize=(10,10))
    # ms.visualise.visualise(list_of_domains[i],color_by='Broad celltype (liver)',objects_to_plot=('collection','Cell boundaries'),shape_kwargs=dict(linewidth=0.1,alpha=1),ax=axes)
    # ms.visualise.visualise(list_of_domains[i],objects_to_plot=('collection','Tumour_Met_Refined'),shape_kwargs=dict(linewidth=2,linestyle='--',facecolor='none'),ax=axes,add_cbar=False,show_boundary=True)
    ms.io.save_domain(list_of_domains[i], name_of_file = names[i]+'.muspan', path_to_save = output_path)
```
