```python
import muspan as ms
import numpy as np  
import matplotlib.pyplot as plt
```

```python

### !!!edit this to the path where you have the data stored!!!
path_to_data = '../data/everything/local_mgn_hack_data/'

# relative paths from the muspan data directory
path_extension_to_4d = 'Muspan_data/AKPT_Liver_Mets/4d/'
path_extension_to_28d = 'Muspan_data/AKPT_Liver_Mets/28d/'
path_extension_primary = 'Muspan_data/AKPT_Primary/'
```

```python
sample_of_interest = 'AKPT_Liver_Mets_28d_1_Region_1.muspan'
domain_28d_1 = ms.io.load_domain(path_to_data+path_extension_to_28d+sample_of_interest)
```

```python
sample_of_interest = 'AKPT_Liver_Mets_4d_1_Region_1.muspan'
domain_4d_1 = ms.io.load_domain(path_to_data+path_extension_to_4d+sample_of_interest)
```

```python
query_tumour_28 = ms.query.query(domain_28d_1, ('label', 'Broad celltype (liver)'), 'is', 'Tumour Epithelial')
```

```python
domain_28d_1.convert_objects(population=query_tumour_28, 
                             collection_name = 'Tumour_Met_New',
                             object_type = 'shape',
                             conversion_method = 'alpha shape',
                             conversion_method_kwargs = dict(alpha = 50))
```

```python
fig,axes=plt.subplots(figsize=(10,10))
ms.visualise.visualise(domain_28d_1,color_by='Broad celltype (liver)',objects_to_plot=('collection','Cell boundaries'),shape_kwargs=dict(linewidth=0.1,alpha=1),ax=axes)
ms.visualise.visualise(domain_28d_1,objects_to_plot=('collection','Tumour_Met_Refined'),shape_kwargs=dict(linewidth=2,linestyle='--',facecolor='none'),ax=axes,add_cbar=False,show_boundary=True)
```

```python
axes.set_xlim([2000,6000])
axes.set_ylim([2000,6000])
```

```python
fig,axes=plt.subplots(figsize=(10,10))
ms.visualise.visualise(domain_28d_1,color_by='Broad celltype (liver)',objects_to_plot=('collection','Cell boundaries'),shape_kwargs=dict(linewidth=0.1,alpha=1),ax=axes)
ms.visualise.visualise(domain_28d_1,objects_to_plot=('collection','Tumour_Met_New'),shape_kwargs=dict(linewidth=1,linestyle='--',facecolor='none'),ax=axes,add_cbar=False,show_boundary=True)
axes.set_xlim([2000,6000])
axes.set_ylim([2000,6000])
```
